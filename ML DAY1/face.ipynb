```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

```python
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread("christhor.jpeg")
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
faces = face_cascade.detectMultiScale(gray)
```

```python
face = faces[0]
```

```python
x,y,w,h = face
```

```python
face = gray[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(gray,(x,y),(x+w,y+h),(255,0,0),4)
```

```python
plt.imshow(out)
```
