```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
bias =100
x,y,coef = make_regression(n_samples =100, n_features =1, noise=10, random_state=42, bias=bias,coef = True)
#bias is intercept(c) in y=mx+c and m is coef or slope
```

```python
x.shape ,y.shape
```

```python
plt.scatter(x,y)
```

```python
y_gen = x * coef + bias
```

```python
plt.scatter(x,y)
plt.plot(x,y_gen)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y) #giving trainging data and model training 
```

```python
model.coef_,model.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,y_gen,label="pre")
plt.plot(x,model.predict(x),label="LR")
plt.legend
```
